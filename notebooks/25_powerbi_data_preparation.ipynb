# Notebook 25 – Power BI Data Preparation

## Objective

This notebook prepares explainability datasets for Power BI.

It converts the wide-format transaction explainability dataset into a normalized long-format dataset suitable for interactive visualizations.

### Input

transaction_explainability.csv

### Output

feature_contributions_long.csv

## Imports

```python
import os
import pandas as pd
```

## Paths

```python
INPUT_FILE = "../powerbi_assets/explainability/transaction_explainability.csv"

OUTPUT_FILE = "../powerbi_assets/explainability/feature_contributions_long.csv"
```

## Load Dataset

```python
df = pd.read_csv(INPUT_FILE)

print(df.shape)

display(df.head())
```

## Convert Wide → Long

```python
records = []

for _, row in df.iterrows():

    for rank in range(1, 6):

        records.append({

            "Transaction_ID": row["Transaction_ID"],

            "Prediction": row["Prediction"],

            "Fraud_Probability": row["Fraud_Probability"],

            "Risk_Level": row["Risk_Level"],

            "Recommendation": row["Recommendation"],

            "Rank": rank,

            "Feature": row[f"Top_Feature_{rank}"],

            "Contribution": row[f"Contribution_{rank}"]

        })

feature_long = pd.DataFrame(records)
```

## Preview

```python
display(feature_long.head(20))
```

## Sort

```python
feature_long = (

    feature_long

    .sort_values(

        ["Transaction_ID", "Rank"]

    )

    .reset_index(drop=True)

)
```

## Export

```python
feature_long.to_csv(

    OUTPUT_FILE,

    index=False

)

print("Power BI Dataset Created Successfully")

print()

print(OUTPUT_FILE)
```
