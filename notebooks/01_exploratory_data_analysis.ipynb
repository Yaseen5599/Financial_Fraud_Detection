```python
import matplotlib.pyplot as plt
import seaborn as sns

# Better looking plots
sns.set_style("whitegrid")

# Increase default figure size
plt.rcParams["figure.figsize"] = (8,5)
```

# Fraud vs Genuine Transactions (Bar Chart)

```python
plt.figure(figsize=(6,4))

sns.countplot(x="Class", data=df)

plt.title("Fraud vs Genuine Transactions")
plt.xlabel("Class (0 = Genuine, 1 = Fraud)")
plt.ylabel("Number of Transactions")
ax = sns.countplot(x="Class", data=df)

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

### Observation

- Genuine transactions (Class = 0) dominate the dataset.
- Fraudulent transactions (Class = 1) are extremely rare.
- This indicates a severe class imbalance problem.
- Therefore, accuracy alone will not be a reliable evaluation metric.
- Precision, Recall, F1-Score, and ROC-AUC will be more appropriate for model evaluation.

# Fraud Percentage (Pie Chart)

```python
fraud_counts = df["Class"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    fraud_counts,
    labels=["Genuine", "Fraud"],
    autopct="%1.2f%%",
    startangle=90,
    explode=(0,0.2)
)

plt.title("Fraud Percentage")

plt.show()
```

### Observation

- Genuine transactions account for approximately 99.83% of all transactions.
- Fraudulent transactions account for only about 0.17%.
- This confirms that fraud detection is an imbalanced classification problem.

# Histogram of Transaction Amount

```python


plt.figure(figsize=(10,5))

plt.title("Histogram of Transaction Amount")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")



sns.histplot(df["Amount"], bins=100)

plt.xlim(0,1000)

plt.show()
```

### Observation

- Most transaction amounts are concentrated at lower values.
- The distribution is highly right-skewed.
- A small number of very large transactions create a long tail.

# Transaction Amount Distribution (Box Plot)

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Amount"])

plt.title("Transaction Amount Distribution")

plt.show()
```

### Observation

- Transaction amounts contain numerous outliers.
- Most transactions fall within a relatively small range.
- A few transactions have exceptionally large values.

# Fraud vs Genuine Transaction Amounts

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="Class", y="Amount", data=df)

plt.title("Transaction Amount by Transaction Type")
plt.xlabel("Class (0 = Genuine, 1 = Fraud)")
plt.ylabel("Transaction Amount")

plt.show()
```

### Observation

- Genuine transactions cover a much wider range of transaction amounts.
- Fraudulent transactions are generally smaller, although some larger frauds are present.
- Transaction amount alone is unlikely to distinguish fraud from genuine transactions.

# Correlation Heatmap

```python
plt.figure(figsize=(18,12))

corr = df.corr()

sns.heatmap(
    corr,
    cmap="coolwarm",
    center=0
)

plt.title("Correlation Heatmap")

sns.heatmap(
    corr,
    cmap="coolwarm",
    center=0,
    linewidths=0.2
)

plt.show()
```

### Observation

- Most variables have weak correlations with one another.
- Some PCA-transformed features show stronger positive or negative relationships with the target variable.
- These correlated features may contribute more to fraud detection models.

## Key Findings

1. The dataset contains 284,807 transactions and 31 features.

2. Fraudulent transactions account for only 0.17% of the dataset, making it a highly imbalanced classification problem.

3. There are no missing values in the dataset.

4. Duplicate transactions exist and should be evaluated before preprocessing.

5. Transaction amounts are highly right-skewed with several extreme outliers.

6. Genuine transactions span a much wider range of transaction amounts than fraudulent transactions.

7. The correlation analysis indicates that some PCA-transformed features have stronger relationships with the target variable than others.

8. Due to severe class imbalance, evaluation metrics such as Precision, Recall, F1-score, and ROC-AUC will be more informative than accuracy.
