## Import Libraries

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
```

## Load Dataset

```python
df = pd.read_csv("../data/raw/creditcard.csv")
```

## Dataset Shape

```python
print(df.shape)
```

## Check Data Types

```python
df.dtypes
```

## Missing Values

```python
df.isnull().sum()
```

## Duplicate Records

```python
duplicates = df.duplicated().sum()

print(duplicates)
```

## Remove Duplicates

```python
df = df.drop_duplicates()
```

### Verify

```python
df.duplicated().sum()
```

## Dataset Shape Again

```python
print(df.shape)
```

## Remove Unnecessary Columns?

### Observation

No columns were removed during preprocessing because all available features may contain predictive information relevant to fraud detection.

## Scale Amount

```python
scaler_amount = StandardScaler()

df["Amount"] = scaler_amount.fit_transform(df[["Amount"]])
```

## Scale Time

```python
scaler_time = StandardScaler()

df["Time"] = scaler_time.fit_transform(df[["Time"]])
```

### Verify

```python
df.head()
```

```python
from pathlib import Path

Path("../data/processed").mkdir(parents=True, exist_ok=True)
```

```python
df.to_csv("../data/processed/creditcard_cleaned.csv", index=False)
```
