## Imports

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

## Create SparkSession

```python
spark = SparkSession.builder \
    .appName("Financial Fraud Detection") \
    .getOrCreate()
```

## Load the Dataset

```python
df = spark.read.csv(
    "../data/raw/creditcard.csv",
    header=True,
    inferSchema=True
)
```

## Verify the Data

```python
df.show(5)
```

## Display the Schema

```python
df.printSchema()
```

## Count Rows

```python
print("Rows:", df.count())
```

## Count Columns

```python
print("Columns:", len(df.columns))
```

## Display Column Names

```python
print(df.columns)
```
