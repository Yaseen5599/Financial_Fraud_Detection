# Model Comparison

## Financial Fraud Detection

This notebook compares the performance of all machine learning models developed for fraud detection.

### Models Compared

- Logistic Regression
- Decision Tree
- Random Forest

### Evaluation Metrics

- Accuracy
- Precision
- Recall
- F1 Score
- ROC-AUC

The evaluation results are loaded from CSV files generated by each model notebook.

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Evaluation Results

```python
lr = pd.read_csv("../results/logistic_regression_metrics.csv")

dt = pd.read_csv("../results/decision_tree_metrics.csv")

rf = pd.read_csv("../results/random_forest_metrics.csv")
```

## Combine Results

```python
results = pd.concat(
    [lr, dt, rf],
    ignore_index=True
)

results
```

## Accuracy Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["Accuracy"])

plt.title("Accuracy Comparison")

plt.xlabel("Model")

plt.ylabel("Accuracy")

plt.ylim(0.9,1.0)

plt.show()
```

## Precision Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["Precision"])

plt.title("Precision Comparison")

plt.xlabel("Model")

plt.ylabel("Precision")

plt.show()
```

## Recall Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["Recall"])

plt.title("Recall Comparison")

plt.xlabel("Model")

plt.ylabel("Recall")

plt.show()
```

## F1 Score Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["F1 Score"])

plt.title("F1 Score Comparison")

plt.xlabel("Model")

plt.ylabel("F1 Score")

plt.show()
```

## ROC-AUC Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["ROC-AUC"])

plt.title("ROC-AUC Comparison")

plt.xlabel("Model")

plt.ylabel("ROC-AUC")

plt.show()
```

## Rank Models

```python
results.sort_values(
    by="F1 Score",
    ascending=False
)
```

## Best Model

```python
best_model = results.sort_values(
    by="F1 Score",
    ascending=False
).iloc[0]

print(best_model)
```

# Final Observation

Three machine learning models were trained and evaluated for financial fraud detection.

## Logistic Regression

- High Recall
- Very Low Precision
- Large number of False Positives

---

## Decision Tree

- Better Precision than Logistic Regression
- Moderate Recall
- Moderate F1 Score
- Lower ROC-AUC

---

## Random Forest

- Highest Precision
- Highest F1 Score
- Highest ROC-AUC
- Excellent overall balance
- Very low False Positives

---

# Best Model

Random Forest achieved the best overall performance among all three models.

Although Logistic Regression detected more fraud cases, it generated many false alarms.

Random Forest achieved an excellent balance between fraud detection capability and prediction reliability, making it the most suitable model for deployment in this project.
