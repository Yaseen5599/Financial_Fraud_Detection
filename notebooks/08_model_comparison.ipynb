# Model Comparison

## Financial Fraud Detection

In this notebook, we compare the performance of all machine learning models trained for fraud detection.

Models Compared:

- Logistic Regression
- Decision Tree
- Random Forest

Evaluation Metrics:

- Accuracy
- Precision
- Recall
- F1 Score
- ROC-AUC

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Create Results Table

```python
results = pd.DataFrame({

    "Model":[
        "Logistic Regression",
        "Decision Tree",
        "Random Forest"
    ],

    "Accuracy":[
        0.9738,
        0.9977,
        0.9994
    ],

    "Precision":[
        0.0534,
        0.3975,
        0.9091
    ],

    "Recall":[
        0.8737,
        0.6737,
        0.7368
    ],

    "F1 Score":[
        0.1006,
        0.5000,
        0.8140
    ],

    "ROC-AUC":[
        0.9556,
        0.8360,
        0.9591
    ]

})

results
```

## Display Results

```python
results
```

## Accuracy Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["Accuracy"])

plt.title("Accuracy Comparison")

plt.ylabel("Accuracy")

plt.ylim(0.9,1.0)

plt.show()
```

## Precision Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["Precision"])

plt.title("Precision Comparison")

plt.ylabel("Precision")

plt.show()
```

## Recall Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["Recall"])

plt.title("Recall Comparison")

plt.ylabel("Recall")

plt.show()
```

## F1 Score Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["F1 Score"])

plt.title("F1 Score Comparison")

plt.ylabel("F1 Score")

plt.show()
```

## ROC-AUC Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["ROC-AUC"])

plt.title("ROC-AUC Comparison")

plt.ylabel("ROC-AUC")

plt.show()
```

## Rank Models

```python
results.sort_values(
    by="F1 Score",
    ascending=False
)
```

# Final Observation

Three machine learning models were evaluated for financial fraud detection.

### Logistic Regression

- Highest Recall
- Very poor Precision
- Large number of False Positives

---

### Decision Tree

- Better Precision than Logistic Regression
- Moderate Recall
- Lower ROC-AUC

---

### Random Forest

- Highest Precision
- Highest F1 Score
- Highest ROC-AUC
- Excellent balance between fraud detection and false alarms

---

## Best Model

Random Forest is selected as the final model because it achieved the best overall performance across all evaluation metrics.

Although Logistic Regression detected more fraud cases, it generated many false positives. Random Forest provides a much better balance between identifying fraud and minimizing incorrect fraud alerts, making it the most suitable model for deployment in this project.
